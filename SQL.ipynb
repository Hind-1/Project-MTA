```python
def get_data(nums):
    dfs=[]
    URL='http://web.mta.info/developers/data/nyct/turnstile/turnstile_{}.txt'
    for i in nums:    
        URLs=URL.format(i)
        dfs.append(pd.read_csv(URLs))        
    return pd.concat(dfs)
nums=[160903,160910,160917]
get_data(nums)
```
